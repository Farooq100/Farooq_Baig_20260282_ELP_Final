## Question 3 [30 Marks] :
Write a program that reads in the `elp_sales.csv` and compute and display the maximum, minimum and average profit for countries in **Australia and Oceania region**. For each country in the Australia and Oceania region, plot their total profit. 

*(Hint: read the CSV into a Pandas dataframe, create a new dataframe for the region, and use the new dataframe for the calculations and bar chart. Compute the total profit before computing the max, min and average)*. 

Save your answer booklet to your GitHub repository.


```python
# Importing pandas library
import pandas as pd

# Reading file from the repository
df = pd.read_csv('/content/elp_sales_data.csv')

# Extracting that specific region of Australia and Oceania from  the dataset and displaying it
df = df.loc[df['Region'] == "Australia and Oceania"]

# Displaying the data
df
```

```python
# Deriving Profit column from the the two columns in order to find min, max and average for the profit
df["Profit"] = df["Total Revenue"] - df["Total Cost"]

# Plotting profit coumn in dataframe
df
```

```python
# FInding Average, Min and Max of the total profit also finding the total profit
print("Maximum profit is",df['Profit'].max())
print("Minimum profit is",df['Profit'].min())
print("Average profit is",df['Profit'].mean())
print("Total profit is",df['Profit'].sum())
```

```python
# Getting each country profit 
pd.unique(df[['Country','Profit']].values.ravel())
```

```python
import matplotlib.pyplot as plt
# For visualisation importing the library

# Getting list of country
lis = df['Country'].unique()

# Net profit of each country
profit = [951410.50, 285087.64, 337937.60, 5270.67,13009.18, 576604.0,727423.20, 146875.14, 1678540.98]


# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(lis, profit)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Graph',
             loc ='left', )
 
plt.xlabel('Profit')
plt.ylabel('Countries')
# Show Plot
plt.show()
```
